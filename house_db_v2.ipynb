```python
import os
import sqlite3
import pandas as pd
import random

# Database filename
db_name = "house_database.db"

# Define dataset path
dataset_path = "dataset"

# Define function to get image files
def get_image_files():
    """Scan dataset folder and extract image file paths along with styles."""
    image_extensions = (".jpg", ".jpeg", ".png", ".bmp", ".gif", ".tiff", ".webp")
    house_data = []
    for root, _, files in os.walk(dataset_path):
        style = os.path.basename(root)  # Get style from folder name
        for file in files:
            if file.lower().endswith(image_extensions):
                image_path = os.path.join(root, file).replace("\\", "/")  # Ensure forward slashes
                house_data.append((image_path, style))
    return house_data

# Create SQLite database
def create_database():
    """Create SQLite database and house table."""
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS houses (
                        id INTEGER PRIMARY KEY AUTOINCREMENT,
                        image_path TEXT,
                        style TEXT,
                        address TEXT,
                        latitude REAL,
                        longitude REAL,
                        area_size REAL,
                        price REAL,
                        bedrooms INTEGER,
                        bathrooms INTEGER,
                        facilities TEXT,
                        magnet TEXT)''')
    conn.commit()
    conn.close()

# Generate random house attributes
bangkok_locations = [
    ("Sukhumvit, Thailand", 13.7245601, 100.5781221),
    ("Silom, Thailand", 13.7278956, 100.5298762),
    ("Sathorn, Thailand", 13.7206225, 100.5258312),
    ("Ari, Thailand", 13.7850576, 100.5445075),
    ("Rama 9, Thailand", 13.7563309, 100.5674509)
]

magnet_options = [
    "Hospital", "School", "Shopping Mall", "Park", "Public Transport", "University",
    "Market", "Office Building", "Restaurant Hub", "Cultural Site"
]

facilities_options = [
    "Pool, Gym, Parking", "Garden, Security", "Balcony, Kitchen", "Smart Home, CCTV"
]

def generate_random_data(image_path, style):
    """Generate random house attributes for each image."""
    address, lat, lon = random.choice(bangkok_locations)
    area_size = round(random.uniform(30, 500), 2)  # Area in sq.m.
    price = round(random.uniform(1_000_000, 50_000_000), 2)  # Price in THB
    bedrooms = random.randint(1, 6)
    bathrooms = random.randint(1, 5)
    facilities = random.choice(facilities_options)
    magnet = ", ".join(random.sample(magnet_options, random.randint(1, 3)))  # Multiple magnet locations
    return (image_path, style, address, lat, lon, area_size, price, bedrooms, bathrooms, facilities, magnet)

# Insert data into SQLite database
def insert_data(house_data):
    """Insert image metadata and generated data into the database."""
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()
    for image_path, style in house_data:
        random_data = generate_random_data(image_path, style)
        cursor.execute('''INSERT INTO houses (image_path, style, address, latitude, longitude, area_size, price, bedrooms, bathrooms, facilities, magnet)
                          VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)''', random_data)
    conn.commit()
    conn.close()

# Read database
def read_database():
    """Read and display the house database."""
    conn = sqlite3.connect(db_name)
    df = pd.read_sql_query("SELECT * FROM houses", conn)
    conn.close()
    return df

# Main execution function
def main():
    """Main function to create and populate the database."""
    print("Scanning image files...")
    house_data = get_image_files()
    
    print("Creating database...")
    create_database()
    
    print("Inserting data...")
    insert_data(house_data)
    
    print("Database setup complete!")
    
    print("Reading database...")
    df = read_database()
    print(df.head())  # Display first few rows

if __name__ == "__main__":
    main()
```
